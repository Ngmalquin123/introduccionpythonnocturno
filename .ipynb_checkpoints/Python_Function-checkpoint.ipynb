### 1 Funciones
1.1 ¿Qué son?
* Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera
que se puedan ejecutar cuando el programador considere oportuno.

* Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
* En cada invocación de la función, los parámetros de entrada pueden variar.

Sabiendo que la ecuación general de segundo grado es x=b+-raiz(b2-4ac)/2*a crear una función donde el usuario mediante teclado ingrese los valores de a,by c calcular las racices de la ecuación

```python
import math

def calc_ecuación_segundo_grado(a, b, c):
    discriminante = b**2 - 4*a*c
    denominador = 2*a

    if denominador == 0:
        return "El valor del denominador es 0 y no se puede dividir para cero"

    if discriminante > 0:
        x1 = (-b + math.sqrt(discriminante)) / denominador
        x2 = (-b - math.sqrt(discriminante)) / denominador
        return x1, x2

    elif discriminante == 0:
        x = -b / denominador
        return x

    else:
        return "El discriminante es negativo, no tiene raíces reales"


a = float(input("Ingrese el valor para a: "))
b = float(input("Ingrese el valor para b: "))
c = float(input("Ingrese el valor para c: "))

calcular = calc_ecuación_segundo_grado(a, b, c)

print(calcular)
```

```python
def sumar(x, y):# Definimos una función sencilla que suma dos números
    return x + y
    
f = sumar# se crea una nueva referencia llamada 'f' que apunta al mismo objeto función  'sumar'.

print(id(f)) #imprime la dirección de memoria (ID único) del objeto f
print(id(sumar))# Imprime la dirección de memoria del objeto 'sumar' y ambos IDs son idénticos porque ambos nombres apuntan al mismo lugar.
print(type(f))# Muestra el tipo de dato de 'f'.
print(sumar(5,6))# Llamada normal usando el nombre original
print(f(5,6))# Llamada usando la  nueva referencia 'f'
```

```python
# Esta función aplica una lista de operaciones a una lista de cadenas
def clean_strings(strings, ops):
    result = []
    for value in strings:        # Para cada ciudad en la lista
        for function in ops:    # Para cada función en la lista de operaciones
            value = function(value) # Aplica la función y actualiza 'value'
            result.append(value)    #  Aquí guardas el valor en CADA paso intermedio
    return result

# Una función personalizada que añade '_5' al final de un texto
def concat5(x):
    return x + '_5'

strings = [' valencia ', ' barcelona', ' bilbao ']

# Crea una lista que contiene 3 funciones:
# 1. str.strip (quita espacios), 2. str.title (pone la primera letra en mayúscula), 3. concat5 (nuestra función)
operations = [str.strip, str.title, concat5]

print(clean_strings(strings, operations))
```

* Si quiero modificar un objeto que es de un tipo básico, los cuales se pasan por valor, puedo
simplemente devolver el resultado de la función y hacer una asignación

```python
# Definimos la función. Recibe un valor, lo multiplica y devuelve el resultado.
def multiplicar_por_2(x):
    # Aquí 'x' es una variable local que recibe el valor 3.
    # Al hacer x * 2, se crea un nuevo objeto (6) y se asigna a 'x'.
    x = x * 2
    # Enviamos el 6 de vuelta al lugar donde se llamó la función.
    return x

# 1. Creamos la variable 'a' y le asignamos el valor 3.
a = 3

# 2. Llamamos a la función pasando 'a'. 
# El resultado (6) sobreescribe el valor antiguo de 'a'.
a = multiplicar_por_2(a)

# 3. Imprimimos el nuevo valor de 'a'.
print(a)
```

* Si quiero que no se modifique un objeto de un tipo complejo, los cuales se pasan por referencia,
puedo pasar una copia a la función:

```python
# Definimos una función que recibe un objeto 'x' (se espera una lista)
def anyadir_2_a(x):
    x.append(2)# El método .append() modifica la lista directamente "in-place"
    print('Dentro de la función: ', x) # Muestra el estado de la lista que vive dentro de la función

lista = [0, 1]# 1. Creamos una lista  con dos elementos

# 2. LLAMADA : No pasamos 'lista', sino 'lista.copy()'
# Esto crea una réplica exacta en una nueva dirección de memoria.
# La función trabaja sobre la copia, no sobre la original.
anyadir_2_a(lista.copy())

# 3. Imprimimos la lista original para ver si sufrió cambios
print('Fuera de la función: ', lista)
```

```python
# Definimos una función que recibe un objeto 'x' 
def anyadir_2_a(x):
    # El método .append() modifica la lista directamente "in-place"
    x.append(2)
    # Muestra el estado de la lista que vive dentro de la función
    print('Dentro de la función: ', x)

# 1. Creamos una lista original con dos elementos
lista = [0, 1]

# 2. LLAMADA CRUCIAL: No pasamos 'lista', sino 'lista.copy()'
# Esto crea una réplica exacta en una nueva dirección de memoria.
# La función trabaja sobre la copia, no sobre la original.
anyadir_2_a(lista.copy())

# 3. Imprimimos la lista original para ver si sufrió cambios
print('Fuera de la función: ', lista)
```

* Si la lista tiene sublistas anidadas hay que hacer un deep copy

```python
import copy

lista = [2, 4, 16, 32, [34, 10, [5, 5]]]

# copy.deepcopy() no solo copia la lista principal, sino que entra
# recursivamente en cada sub-lista y crea copias nuevas de TODO.
copia = copy.deepcopy(lista)

# Modificamos un valor simple en la copia
copia[0] = 454

# Modificamos un valor profundamente anidado: el primer 5 de [5, 5]
copia[4][2][0] = 64

print(lista) # Se mantiene intacta: [2, 4, 16, 32, [34, 10, [5, 5]]]
print(copia) # Refleja los cambios: [454, 4, 16, 32, [34, 10, [64, 5]]]
```

* Reasignar un parámetro nunca afecta al objeto de fuera:

```python
def funcion_poco_util(x):
    # 1. Al entrar, 'x' apunta al mismo objeto que 'lista'. Los IDs coinciden.
    print(id(x))
    
    # Al usar 'x = [2, 3]', le dices a 'x' que deje de mirar a la lista original
    # y empiece a mirar a una nueva lista en otra dirección de memoria.
    x = [2, 3]
    
    # 3. Modifica la nueva lista
    x.append(4)
    
    # 4. Este id será totalmente diferente al primero.
    print(id(x))

lista = [0, 1]
# Imprime el ID de la lista original.
print(id(lista))

# Llamamos a la función.
funcion_poco_util(lista)

# Comprobamos la lista original.
print(lista)
```

### 1.6 Argumentos arbitrarios
* No se sabe a priori cuantos elementos se reciben.
* Se reciben como una tupla.
* Los parámetros se especifican con el símbolo ’*’.

```python
# El '*' permite que la función reciba cualquier número de argumentos (0, 1, 10...)
def saludar(*names):
    # Python guarda todos los argumentos recibidos dentro de una TUPLA
    print(type(names))
    
    # Recorremos la tupla para saludar a cada persona
    for name in names:
        print("Hola " + name)

# Llamada 1: Sin argumentos
# 'names' será una tupla vacía: (). No imprimirá ningún hola.
saludar()

# Llamada 2: Con cinco argumentos
# Python los empaqueta todos en una tupla: ('Manolo', 'Pepe', 'Luis', 'Alex', 'Juan')
saludar("Manolo", "Pepe", "Luis", "Alex", "Juan")
```

```python
 # El '**' indica que la función recibirá un número variable de argumentos
# con nombre (como clave=valor) y los guardará en un DICCIONARIO llamado 'kwargs'.
def consulta_sql(**kwargs):
    # Imprime el diccionario generado: {'station': 'alameda', 'free': 10}
    print(kwargs)
    
    # Base de la sentencia SQL
    sql = "SELECT * FROM bicis WHERE "
     
    # Recorre el diccionario y crea una lista de strings con formato "llave=valor".
    # Resultado: ["station=alameda", "free=10"]
    args = [f"{key}={value}" for key, value in kwargs.items()]
    
    # Une todos los elementos de la lista 'args' usando el separador " AND ".
    # Luego lo concatena a la base de la variable 'sql'.
    sql += " AND ".join((args))
    
    return sql

# Llamada a la función pasando dos argumentos con nombre
print(consulta_sql(station='alameda', free=10))
```

* 1.6.1 Desempaquetado en funciones

```python
# Definimos una función que requiere exactamente 3 parámetros: 'quien', 'mensaje' y 'gesto'
def saludar(quien, mensaje, gesto):
    # Usamos una f-string para imprimir los valores de forma limpia
    print(f"Hola {quien}, {mensaje}, mira esto: {gesto}")

# Creamos una lista que contiene exactamente 3 elementos
ls_saludo = ['Manolo', 'te quiere saludar', 'args']

# FORMA 1: Manual (Posicional)
# Accedemos a cada índice de la lista (0, 1, 2) y los pasamos individualmente.
# Es funcional, pero más largo de escribir.
saludar(ls_saludo[0], ls_saludo[1], ls_saludo[2])

# FORMA 2: Desempaquetado con '*'
# El '*' le dice a Python: "Toma esta lista, ábrela y entrega cada elemento 
# a la función en el orden en que aparecen".
saludar(*ls_saludo)
```

```python
# La función espera tres parámetros con nombres específicos
def saludar(quien, mensaje, gesto):
    print(f"Hola {quien}, {mensaje}, mira esto: {gesto}")

# Creamos un diccionario donde las claves coinciden con los nombres de los parámetros
dc_saludo = {
    'quien': [12, 12, 3, 4, 5, 3], # El valor puede ser cualquier cosa, incluso una lista
    'gesto': 'guiño',
    'mensaje': 'te quiere enseñar esto'
}

# El '**' le dice a Python: "Busca en este diccionario y pasa cada valor 
# al parámetro de la función que tenga el mismo nombre que la clave".
saludar(**dc_saludo)
```

### 1.7 Recursividad
* Funciones que se llaman a sí mismas

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(10))

def fact(n):
    a = 1
    for i in range(1, n+1):
        a *= i
    return a
print(fact(10))
```

```python
fact = lambda x: 1 if x == 0 else x * fact(x-1)
fact(10)
```

### 1.8 Funciones Lambda
* Es posible definir funciones anónimas (sin nombre).
* Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia def
no puede (por ejemplo, dentro de una lista o como parámetro de una función).
* Útiles cuando se quiere pasar una función como argumento a otra.
Formato general:
lambda <lista de argumentos> : <valor a retornar>
* Importante: <valor a retornar> no es un conjunto de instrucciones. Es simplemente una
expresión return que omite esta palabra.
* Las funciones definidas con def son más generales:<br>
– Cualquier cosa que implementes en un lambda, lo puedes implementar como una función
convencional con def, pero no viceversa.

```python
# Ejemplo: función suma.
def suma(x, y, z):
    return x + y + z
print(suma(1, 2, 3))
# Ejemplo: función suma como expresión lambda.
suma2 = lambda x, y, z : x + y + z
print(suma2(1, 2, 3))
```

```python
# La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.
9
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']
cities.sort()
print(cities)
cities.sort(key = lambda x : len(x))
print(cities)
cities.sort(key = len)
print(cities)
def longitud(x):
    return len(x)
cities.sort(key = longitud)
print(cities)
```

```python
# Mismo ejemplo sin lambda
# 1. Definimos una lista de ciudades
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# 2. Creamos una función que toma un valor (x) y devuelve un número (su longitud)
def num_caracteres(x):
    return len(x)

# Llamamos al método .sort()
# El argumento 'key' espera una función. Python aplicará 'num_caracteres'
# a cada elemento de la lista para decidir su posición.
cities.sort(key = num_caracteres)

# 4. Imprimimos el resultado
print(cities)
```

* Nos podemos guardar una referencia para utilización repetida.

```python
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# Definimos una función lambda
# 'x' es el argumento de entrada
# 'len(x)' es lo que la función devuelve automáticamente
f = lambda x : len(x)

# Recorremos la lista de ciudades
for s in cities:
    # Llamamos a la lambda 'f' pasando cada ciudad 's'
    print(f(s))
```

 * Las expresiones lambda pueden formar parte de una lista.

```python
#Lista de lambdas para mostrar las potencias de 2.
pows = [lambda x: x ** 0,
lambda x: x ** 1,
lambda x: x ** 2,
lambda x: x ** 3,
lambda x: x ** 4]
for f in pows:
    print(f(2))
```

* Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
* • Ideal para escoger entre varias opciones

```python
def switch_dict(operator, x, y):
    # crea un diccionario de opciones y selecciona 'operator'
    return {
    'add': lambda: x + y,
    'sub': lambda: x - y,
    'mul': lambda: x * y,
    'div': lambda: x / y,
    }.get(operator, lambda: None)() # retorna None si no encuentra 'operator'
print(switch_dict('add',2,2))
print(switch_dict('div',10,2))
print(switch_dict('mod',17,3))
```

### 1.9 Scopes
* La localización de una asignación a una variable en el código determina desde donde puedes
acceder a esa variable.
* Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
* El alcance de la visibilidad de una variable determina su scope.
* El término scope hace referencia a un espacio de nombres (namespace). Por ejemplo, una

```python
 X = 10
 def func():
    X = 20 # Local a la función. Es una variable diferente, no visible fuera de 'func'
    def func_int():
        X = 30
        print(X)
    func_int()
func()
```

### Regla LEGB 
Dada una función F, tenemos los siguientes scopes:
* Local: variables locales a F.
* Enclosing: variables localizadas en funciones que contienen a (o por encima de) F.
* Global: variables definidas en el módulo (fichero) que no están contenidas en ninguna función.
Este scope no abarca más de un fichero.
* Built-ins: proporcionadas por el lenguaje.

La resolución de nombres ocurre de abajo a arriba.

```python
#Global (X y func)
X = 99
def func(Y): # Local (Y y Z)
    Z = X + Y
    return Z
print(func(1))
```

```python
X = 1
def func():
    X = 2 # X es una variable diferente
func()
print(X)
```

* La sentencia global nos permite modificar una variable global desde dentro de una función.

```python
X = 1
def func():
    global X
    X = 2
func()
print(X)
```

* No hay necesidad de usar global para referenciar variables. Sólo es necesario para modificarlas.

```python
y, z = 1, 2  # Se crean dos variables en el ámbito GLOBAL.

def todas_globales():
    # 'global x' le dice a Python: "No crees una 'x' solo para esta función.
    # Si 'x' no existe fuera, créala en el espacio global".
    global x 
    
    # Python busca 'y' y 'z'. Como no están dentro de la función,
    # las busca fuera (en el ámbito global) y las encuentra.
    x = y + z 

todas_globales()

# Ahora podemos acceder a 'x' incluso fuera de la función 
# porque fue declarada como global.
print(x)  # Resultado: 3
print(y)  # Resultado: 1
print(z)  # Resultado: 2
```

```python
X = 77  # Ámbito GLOBAL

def f1():
    X = 88  # Ámbito ENCLOSING (local para f1, pero externo para f2)
    
    def f2():
        # Aquí ocurre la búsqueda:
        # 1. ¿Hay una 'X' dentro de f2 (Local)? NO.
        # 2. ¿Hay una 'X' en la función que envuelve a f2 (Enclosing)? SÍ.
        print(X) 
        
    f2()

f1()
```

### Closures
* Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan
existiendo o no.

```python
def f1():
    X = 88  # Variable local de f1
    
    def f2():
        # f2 "atrapa" la variable X de su entorno
        print(X)
    
    # IMPORTANTE: Devolvemos la función f2 como si fuera un objeto
    # No estamos haciendo f2(), estamos enviando la "receta" de la función.
    return f2 

# 1. Llamamos a f1. Al terminar, la variable local 'X' debería desaparecer de la memoria...
# ... pero como f1 devuelve f2, y f2 necesita a X, Python mantiene viva esa variable.
accion = f1()

# 2. Ahora 'accion' es la función f2. 
# 3. La ejecutamos fuera de su contexto original.
accion()
```

* Sentencia nonlocal para modificar variables que no son locales, pero tampoco globales.

```python
def f1():
    contador = 10
    
    def f2():
        # 'nonlocal' le dice a Python: "No crees una variable nueva.
        # Usa la variable 'contador' que pertenece a f1".
        nonlocal contador
        contador += 1
        print(contador)
    
    return f2

# 'accion' ahora es la función f2, llevando consigo su propio 
# "maletín" con la variable 'contador' inicializada en 10.
accion = f1()

accion() # Incrementa a 11 y lo imprime
accion() # Incrementa a 12 y lo imprime
accion() # Incrementa a 13 y lo imprime
```

5. Escribe una función que compruebe si un número se encuentra dentro de un rango específico.

6. Escribe una función que reciba un número entero positivo como parámetro y devuelva una
lista que contenga los 5 primeros múltiplos de dicho número. Por ejemplo, si la función recibe
el número 3, devolverá la lista [3, 6, 9, 12, 15]. Si la función recibe un parámetro incorrecto

reciba una lista como parámetro y compruebe si la lista tiene duplicados. La función devolverá

9. Escribe una función que compruebe si un string dado es un palíndromo. Un palíndromo es
una secuencia de caracteres que se lee igual de izquierda a derecha que de derecha a izquierda.
Por ejemplo, la función devolverá True si recibe el string “reconocer” y False si recibe el string
“python”.

### Solución

```python
#Ejercicio 1
def cuadrados(lista):
    return [n ** 2 for n in lista]

# Prueba
print(cuadrados([1, 2, 3, 4])) # [1, 4, 9, 16]
```

```python
#Ejercicio 2
def suma_indeterminada(*numeros):
    return sum(numeros)

# Prueba
print(suma_indeterminada(5, 10, 2, 3)) # 20
```

```python
#Ejercicio 3
def invertir_string(texto):
    return texto[::-1]

print(invertir_string("hola")) 
```

```python
#Ejercicio 4
invertir_lambda = lambda texto: texto[::-1]

print(invertir_lambda("mundo")) 
```

```python
#Ejercicio 5
def en_rango(num, inicio, fin):
    return inicio <= num <= fin

# Prueba
print(en_rango(5, 1, 10)) # True
print(en_rango(15, 1, 10)) # False
```

```python
#Ejercicio 6
def cinco_multiplos(n):
    if n <= 0:
        print("Error: El número debe ser mayor que cero.")
        return []
    # Generamos desde n hasta n*5, saltando de n en n
    return list(range(n, (n * 5) + 1, n))

# Prueba
print(cinco_multiplos(3)) # [3, 6, 9, 12, 15]
print(cinco_multiplos(-1)) # Error... []
```

```python
#Ejercicio 7
def tiene_duplicados(lista):
    return len(lista) != len(set(lista))

# Prueba
print(tiene_duplicados([1, 2, 3, 1])) # True
```

```python
#Ejercicio 8
duplicados_lambda = lambda lista: len(lista) != len(set(lista))

# Prueba
print(duplicados_lambda([1, 2, 3])) # False
```

```python
#Ejercicio 9
def es_palindromo(texto):
    # Convertimos a minúsculas para que 'Reconocer' también sea True
    texto = texto.lower()
    return texto == texto[::-1]

# Prueba
print(es_palindromo("reconocer")) # True
print(es_palindromo("python"))    # False
```
