### Bucle ‘while’
* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta
que una condición evalue a False).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.<br>
Formato general:<br>
while test:       # Mientras se cumple la condición<br>
   statements         # Instrucciones a ejecutar

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
```

```python
while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

### 3.1 Las sentencias break, continue y else
#### Break y continue
* Solo tienen sentido dentro de bucles.
* Break permite terminar el bucle por completo.
* Continue permite saltar a la siguiente iteración, continuando con el bucle.
* Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sen-
tencias condicionales (if).

```python
# # Ejemplo: break
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]
for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue # Si el número es par, salta a la siguiente iteración
    print(f'Odd number {i}')
```

* Observa como la sentencia continue te puede ayudar a reducir el número de niveles de
anidamiento.
* Sin continue el anterior ejemplo sería:

```python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
```

```python
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
```

### Else
* Los bucles pueden tener una sensencia else.
* Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros
lenguajes.
* Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de
un break.

```python
lista = [60,60,30,60]
element_to_find = 60
for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
```

```python
# with flag
lista = [10,30,50,40]
element_to_find = 30
found = False
for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
```

### 4 List Comprehensions
* Permiten construir listas a través de la ejecución repetida (sentencia for) de una expressión
para cada item de un objeto iterable.
* Van entre ‘[’ y ’]’. Esto es indicativo de que estamos construyendo una lista.<br>
Sintaxis:<br>
\[\<expression> for \<item> in \<iterable>]<br>

Ejemplo: repetir los carácteres de un string.

```python
[char*3 for char in "Enrique"]
```

```python
lista = []
for char in 'Enrique':
    lista.append(char)
    print(lista)
```

Comúnmente, el item de la sentencia for aparecerá en la expresión principal, pero eso no es obligatorio.

```python
[2 for _ in 'Enrique']
```

### 4.0.1 Versión extendida
Se puede especificar un filtro (sentencia if ) para obtener únicamente los elementos que cumplan
cierta condición.<br>
Sintaxis:<br>
[\<expression> for \<item> in \<iterable> if \<condition>]<br>

Ejemplo: obtener números pares:

```python
[y for y in range(9) if y % 2 == 0]
```

* Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle
equivalente.

```python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
```

### 4.0.2 Versión completa
La sentencia if de una list comprehension también puede contener una expresión alternativa.Sintaxis:
[\<expression_1> if \<condition> else \<expression_2> for \<item> in \<iterable>]<br>
 
Ejemplo: poner a cero los números pares.


```python
[0 if x % 2 == 0 else x for x in range(9)]
```

```python
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
```

```python
lista = []
for x in range(-4,5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
print(lista)
```

```python
[x**2 if x > 0 else 100 if x==0 else x for x in range(-4,5)]
```

```python
lista = []
for x in range(-4,5):
    if x > 0:
        lista.append(x**2)
    elif x == 0:
        lista.append(100)
    else:
        lista.append(x)
print(lista)
```

### 4.0.3 Anidamiento
Las list comprehensions soportan anidamiento en sus expresiones.<br>

Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas
dadas.

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]
[(x, y) for x in lista_1 for y in lista_2]
```

### 4.0.4 Pros y contras

### 4.0.5 Otros tipos de comprehensions
#### Dictionary comprehensions
Usando ‘{’ y ‘}’ como delimitadores y una expresión ‘clave : valor’, se obtiene un diccionario en
lugar de una lista.<br>

Ejemplo: diccionario donde cada valor es el cuadrado de la clave.

```python
d = {x : x*x for x in range(10)}
print(d)
print(type(d))
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
shopping = {k:v for k,v in zip(items,price)}
print(shopping)
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
for k in zip(items, price):
    print(type(k))
    print(k)
```

### 5 Excepciones
* Las excepciones son eventos que representan situaciones excepcionales.
11
* Alteran el flujo de ejecución convencional.
* Python lanza excepciones automáticamente cuando se producen errores.
* El programador puede lanzar excepciones de manera explícita y también capturar excepciones

### Try/except
* Permite capturar excepciones y actuar en consecuencia.<br>
  
Ejemplo: error de acceso fuera de rango.

```python
 lista = [6, 1, 0, 5]
 lista[4]
 print('Código tras el error')
```

```python
 lista = [6, 1, 0, 5]
 i = 300
 try:
     a = lista[i]
 except IndexError:
         print('He capturado la excepción de tipo IndexError')
         a = 0
 print('Código tras el bloque try')
 print(a)
```

 * Normalmente, al capturar excepciones, querremos ser lo más específicos posible, pero también
se puedes usar una sentencia try-except que capture cualquier error.

```python
 try:
     4/0
 except:
    print('He capturado la división por cero')
```

 * Recoger la excepción e imprimir el mensaje de error

```python
 lista = [6, 1, 0, 5]
 try:
     print(lista[4])
 except Exception as e:
     print("Error = " + str(e))
 print('Reacheable Code')
```

### Try/finally
* A través de finally podemos especificar código que queremos que se ejecute siempre (indepen-
dientemente de si se produce la excepción o no).
* Se suele usar para liberar recursos.

```python
 lista = [6, 1, 0, 5]
 try:
     a = lista[1]
 except IndexError:
     print('Exception IndexError Captured')
 finally:
     print('Bloque finally')
     b = lista[2]
 print('Código tras el bloque try')
 print(b)
```

### raise
* La sentencia raise nos permite lanzar excepciones de manera explícita.

```python
 lista = [6, 1, 0, 5]
 indice = 4
 try:
     if indice >= len(lista):
         raise IndexError('Índice fuera de rango')
 except IndexError:
     print('Excepción de tipo IndexError capturada')
```

```python
#ejercicio 1
lista = [2, 5, 7, 10]
suma = 0

for num in lista:
    suma += num

print("La suma es:", suma)
```

```python
#ejercicio 2
lista = [1,2,2,3,4,4,5]
nueva = []

for elemento in lista:
    if elemento not in nueva:
        nueva.append(elemento)

print(nueva)
```

```python
#ejercicio 3
n = 5
diccionario = {}

for x in range(1, n+1):
    diccionario[x] = x*x

print(diccionario)
```

```python
#ejercicio 4
palabras = ["computadora", "automovilistica", "arbol", "python"]

encontrada = False

for p in palabras:
    if p.startswith("a") and len(p) > 9:
        print("Palabra encontrada:", p)
        encontrada = True
        break

if not encontrada:
    print("No se encontró ninguna palabra")
```

```python
#ejercicio 5
L = [3,5,13,12,1,9]
indices = []

for i in range(len(L)):
    if L[i] % 3 == 0:
        indices.append(i)

print(indices)
```

```python
#ejercicio 6
dic = {'a':4.3,'b':1,'c':7.8,'d':-5}

clave_max = max(dic, key=dic.get)

print("Clave con mayor valor:", clave_max)
```

```python
#ejericio 7
a = [2,4,6,8]
b = [7,11,15,22]

for x in a:
    if x > 5:
        for y in b:
            if y < 14:
                print(x * y)
```

```python
#ejercicio 8
try:
    X = float(input("Ingrese un número: "))
    resultado = 10 / X
    print("Resultado:", resultado)

except ZeroDivisionError:
    print("No se puede dividir para cero")

except ValueError:
    print("Debe ingresar un número válido")
```

```python
# ejerciio 9
dic = {"nombre":"Luis","edad":25,"ciudad":"Quito"}

clave = input("Ingrese una clave: ")

try:
    print("Valor:", dic[clave])
except KeyError:
    print("La clave no existe")
```

```python
#ejercicio 10
lista = [1, -2, 3.5, 4, 7.2, 6]

resultado = [x for x in lista if isinstance(x,int) and x > 0]

print(resultado)
```

```python
#ejercicio 11
palabra = "programacion"

vocales = {letra for letra in palabra if letra in "aeiou"}

print(vocales)
```

```python
#ejercicio 12
lista = [x for x in range(51) if "3" in str(x)]

print(lista)
```

```python
#ejercicio 13
frase = "Soy grandioso"

dic = {palabra: len(palabra) for palabra in frase.split()}

print(dic)
```

```python
#ejercicio 14
numeros_texto = ["seis","siete","ocho","nueve","diez"]

lista = [x for x in range(1,6)] + [numeros_texto[i] for i in range(5)]

print(lista)
```
