1. Escribe un programa que pida el peso (en kg) y la estatura (en metros) y calcule el IMC con la fórmula: Los datos de peso y altura deben ser ingresados por teclado

$$IMC=\frac{Peso}{estatura^2}$$
	​


```python
import math 
```

```python
#solución

peso=float(input("ingrese el peso en kilogramos"))
estatura=float(input("ingrese la altura en metros"))
           
imc=peso/(estatura**2)
print("EL IMC ES : ",round(imc,3))
```

Escribe un programa que pida las coordenadas de dos puntos en el plano:
$$(x_1,y_1) y (x_2 , y_2)$$

Y calcule la distacia entre ellos
$$\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}$$

Los datos x1,x2,y2,y2 son ingresados por teclado. Pedir al profesor ayuda para utilizar la función sqrt de la raiz


```python
#solución
x1=float(input("ingrese las cordenadas de x1 del primer punto"))
y1=float(input("ingrese las cordenadas de y1 del primer punto"))
x2=float(input("ingrese las cordenadas de x2 del segundo punto"))
y2=float(input("ingrese las cordenadas de y2 del segundo punto"))
distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("la distancia es: ", round(distancia,3))
```
