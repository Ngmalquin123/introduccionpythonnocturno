# Comentarios

### ¿Qué son?
Texto contenido en ficheros de Python que es ignorado por el intérprete; es decir, no es ejecutado

### ¿Cuál es su utilidad?
* Se puede utilizar para documentarcódigo y haerlo más legible

* Preferebilemente, trataremos de hacer código fácil de entender y que nececite pocos comentarios, en lugar de vernos forzados a recurrir a los comentarios para explicar el código.

### Tipos de comentarios

**Comentarios de una línea**

* Texto precedido por '#'
* Se suele usar para documentar expreciones sencillas


```python
#Esto es una instrucción print
print("Hello word") #Esto es una instrucción print
```

**Comentarios de varias líneas**

*Texto encapsulado en triple comillas (que pueden se simples como doble)

*Se suele usar para documentar bloques de código más significativo


**Función**

```python
# creo una función de multiplicación
def producto(x,y):
    '''
    Esta funcion recibe dos números como parámetros y devuelve
    como resultado el producto de los mismos
    '''
    #creo una variable donde guardo el producto de los números recibidos
    c= x*y
    return c    
```

```python
datos=producto(10,5)
print(datos)
```

# Literales, variables y tipos básicos

De forma muy génerica, al ejecutarse un programa en python, simplemente se realizan dos cosas 
*operaciones* sobre *objetos*

Estos dos términos son fundamentales.

* *Objetos*: Cualquier tipo de datos (númericos, caracteres o datos más complejos).

* *Operaciones*: Cómo manipulamos estos datos. +, -, *,/

  Ejemplo: 

```python
4.3*3
```

## Literales

* Python tiene una serie de tipos de datos integrados en el propio lenguaje
* Los literales son expreciones que generan objetos de estos tipos.
* Estos objetos, según su tipo pueden ser:
  * Simples o compuestos
  * Mutables o inmutables

### Literales simples

- Enteros
- Decimales o puntos flotantes
- Booleanos

```python
print(4) # número entero
print(4.3) # número flotante
print('Hello word!') # objeto de tipo string
print(False)
```
